```python
import pandas as pd

df = pd.read_csv('wordslist.csv', header=None, names=['WORDS'])
df
```

```python
# Add a new column with the length of each word
df['Word_Length'] = df['WORDS'].astype(str).apply(len)

# Create a pivot table with Word_Length as columns
df_pivoted = df.pivot_table(index=df.index, columns='Word_Length', values='WORDS', aggfunc='first')

# Display the pivoted DataFrame without NaN values
df=(df_pivoted.fillna('').astype(str))
```
